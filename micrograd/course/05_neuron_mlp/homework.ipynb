## 完整例子 - 一个神经元就是点积加 bias

`Neuron(2)` 接收两个输入，内部有两个权重和一个 bias，输出一个数。

```python
w = [0.2, -0.5]
x = [2.0, -3.0]
b = 0.1
act = sum((wi*xi for wi, xi in zip(w, x)), b)
print('w0*x0 + w1*x1 + b =', act)
```
