```python
from pathlib import Path
import sys, math, random


def find_repo_root():
    p = Path.cwd().resolve()
    for q in [p, *p.parents]:
        if (q / 'micrograd' / 'engine.py').exists():
            return q
    return p

ROOT = find_repo_root()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))


def near(a, b, eps=1e-6):
    return abs(a - b) < eps


def todo_guard(values, message='请先填写 TODO，再运行检查。'):
    if any(v is None for v in values):
        print(message)
        return False
    return True
```

## 完整例子 - 乘法的局部梯度

`L=a*b` 时，如果固定 `b`，`a` 每增加一点，`L` 大约增加 `b` 倍。

```python
a = 2.0
b = 3.0
h = 0.0001
L1 = a * b
L2 = (a + h) * b
print('数值估计 dL/da =', round((L2 - L1) / h, 4))
print('手算 dL/da = b =', b)
```

## Modify - 参考乘法，换成 `L=a*b+a`

只比完整例子多了一个 `+a`，所以对 `a` 的影响也要多一条路径。

```python
# a=2,b=3, L=a*b+a
student_modify_dL_da = None
student_modify_dL_db = None


def qa_check_01_modify():
    if not todo_guard([student_modify_dL_da, student_modify_dL_db]): return
    assert student_modify_dL_da == 4, 'd(ab+a)/da = b+1 = 4。'
    assert student_modify_dL_db == 2, 'd(ab+a)/db = a = 2。'
    print('OK: Modify 通过。')

qa_check_01_modify()
```

```python
# 请填手算结果。默认 a=2,b=3,c=-4,w=3。
student_grad_L_eq_a_plus_b = None       # d(a+b)/da
student_grad_L_eq_a_times_b_da = None   # d(a*b)/da
student_grad_L_eq_a_times_b_db = None   # d(a*b)/db
student_grad_L_eq_a_square = None       # d(a*a)/da, a=2
student_grad_L_eq_2ab_plus_2a_da = None # d(2(ab+a))/da, a=2,b=3
student_grad_L_eq_2ab_plus_2a_db = None # d(2(ab+a))/db, a=2,b=3
student_grad_L_eq_2w_plus_1_square = None # d((2w+1)^2)/dw, w=3


def qa_check_gradient_practice():
    values = [
        student_grad_L_eq_a_plus_b,
        student_grad_L_eq_a_times_b_da,
        student_grad_L_eq_a_times_b_db,
        student_grad_L_eq_a_square,
        student_grad_L_eq_2ab_plus_2a_da,
        student_grad_L_eq_2ab_plus_2a_db,
        student_grad_L_eq_2w_plus_1_square,
    ]
    if not todo_guard(values): return
    assert student_grad_L_eq_a_plus_b == 1
    assert student_grad_L_eq_a_times_b_da == 3
    assert student_grad_L_eq_a_times_b_db == 2
    assert student_grad_L_eq_a_square == 4, 'a*a 对 a 有两条路径，结果 2a=4。'
    assert student_grad_L_eq_2ab_plus_2a_da == 8
    assert student_grad_L_eq_2ab_plus_2a_db == 4
    assert student_grad_L_eq_2w_plus_1_square == 28
    print('OK: 梯度手算通过。')

qa_check_gradient_practice()
```
