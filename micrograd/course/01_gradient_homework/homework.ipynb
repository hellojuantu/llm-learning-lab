## 完整例子 - 乘法的局部梯度

`L=a*b` 时，如果固定 `b`，`a` 每增加一点，`L` 大约增加 `b` 倍。

```python
a = 2.0
b = 3.0
h = 0.0001
L1 = a * b
L2 = (a + h) * b
print('数值估计 dL/da =', round((L2 - L1) / h, 4))
print('手算 dL/da = b =', b)
```
