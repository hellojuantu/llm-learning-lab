```python
from pathlib import Path
import sys, math, random


def find_repo_root():
    p = Path.cwd().resolve()
    for q in [p, *p.parents]:
        if (q / 'micrograd' / 'engine.py').exists():
            return q
    return p

ROOT = find_repo_root()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))


def near(a, b, eps=1e-6):
    return abs(a - b) < eps


def todo_guard(values, message='请先填写 TODO，再运行检查。'):
    if any(v is None for v in values):
        print(message)
        return False
    return True
```

```python
# L = a*b, a=2, b=3
# TODO: dL/da 和 dL/db 分别是多少？
predict_dL_da = None
predict_dL_db = None

def qa_check_preview_01(da, db):
    if not todo_guard([da, db]): return
    assert da == 3, 'd(a*b)/da = b。'
    assert db == 2, 'd(a*b)/db = a。'
    print('OK: 乘法局部梯度通过。')

qa_check_preview_01(predict_dL_da, predict_dL_db)
```
