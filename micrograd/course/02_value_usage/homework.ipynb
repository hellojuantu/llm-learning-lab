## 完整例子 - 先看 data，再看 grad

同一个表达式，`data` 来自前向计算，`grad` 来自 backward 后的反向传播。

```python
from micrograd.engine import Value

ex_a = Value(2.0)
ex_b = Value(3.0)
ex_L = (ex_a * ex_b + ex_a) * 2
print('before backward:', ex_L, ex_a, ex_b)
ex_L.backward()
print('after backward:', ex_L, ex_a, ex_b)
```
