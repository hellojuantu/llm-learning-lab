## 第一个最小可运行例子

下面这段不是神经网络，只是一个数值表达式。你先观察：`data` 是前向值，`grad` 是 backward 后的导数。

````python
from micrograd.engine import Value

a = Value(2.0)
b = Value(3.0)
L = (a * b + a) * 2
print('before backward:', 'L.data =', L.data, 'a.grad =', a.grad, 'b.grad =', b.grad)
L.backward()
print('after backward:', 'L.data =', L.data, 'a.grad =', a.grad, 'b.grad =', b.grad)
````
