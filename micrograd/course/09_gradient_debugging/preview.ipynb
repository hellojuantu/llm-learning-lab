```python
def student_zero_grad_demo():
    a = Value(3.0)
    L = a * 2
    L.backward()
    first_grad = a.grad
    L.backward()
    second_grad = a.grad
    a.grad = 0
    L.backward()
    after_zero_grad = a.grad
    return first_grad, second_grad, after_zero_grad
```
