## 完整例子 - 先用代码看斜率

先不要背导数。我们用两个很近的点算 `(y2-y1)/(x2-x1)`。

```python
def f_example(x):
    return 2*x + 1

x0 = 3.0
h = 0.0001
slope_example = (f_example(x0 + h) - f_example(x0)) / h
print('y=2x+1 在 x=3 附近的斜率约等于:', round(slope_example, 4))
```
