```python
from pathlib import Path
import sys, math, random


def find_repo_root():
    p = Path.cwd().resolve()
    for q in [p, *p.parents]:
        if (q / 'micrograd' / 'engine.py').exists():
            return q
    return p

ROOT = find_repo_root()
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))


def near(a, b, eps=1e-6):
    return abs(a - b) < eps


def todo_guard(values, message='请先填写 TODO，再运行检查。'):
    if any(v is None for v in values):
        print(message)
        return False
    return True
```

```python
# y 是真实标签，只能是 1 或 -1。
# score 是模型输出，正数偏向 1，负数偏向 -1。
# TODO: y=1, score=0.2 时 margin 是多少？ y=-1, score=-2 时 margin 是多少？
margin_a = None
margin_b = None

def qa_check_preview_07(a, b):
    if not todo_guard([a, b]): return
    assert near(a, 0.2)
    assert near(b, 2.0)
    print('OK: y*score 会把答对方向统一成正数。')

qa_check_preview_07(margin_a, margin_b)
```
